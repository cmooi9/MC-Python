Create a program where the user can play a game of rock, paper, scissors against a computer. Set a play choice (rock, paper, or scissors) for the computer player and during the program, request for the user to input their play choice.

The rules in Rock, Paper, Scissors are:

Rock beats scissors
Scissors beats paper
Paper beats rock
If the user wins, then display "You win!" If they lose against the computer, then display "You lose." If the computer and the user have the same play choice, display "It's a draw."

Also ask the user if they would like to play again. If they do, start the game again but if not, thank the user for playing the game.

Bonus 1: Do not allow the user to select/enter faulty inputs (anything other than rock, paper, scissors). 
Bonus 2: Modify your program so that a point is awarded to whoever wins each round (in case of a tie, do not award points). At the end of each set, show the final score.  Optional: Show the round history at the end. 

```python
import random as rand
computer_choices = ['Rock', 'Paper', 'Scissors']

print('LETS PLAY ROCK, PAPER, SCISSORS')
print()
x = 'yes'
computer_score = 0
player_score = 0

while x != 'no':
    player_inp = input('Rock, Paper, Scissors, Shoot: ')
    player = player_inp.lower()
    computer = rand.choice(computer_choices)
    print(f'Computer: {computer}')
#player rock outcomes
    if player == 'rock' and computer == 'Rock':
        print("It's a draw.")
    elif player == 'rock' and computer == 'Scissors':
        print('You lose.')
        computer_score = computer_score + 1
    elif player == 'rock' and computer == 'Paper':
        print('You win!')
        player_score = player_score + 1

#player scissors outcomes
    elif player == 'scissors' and computer == 'Scissors':
        print("It's a draw.")
    elif player == 'scissors' and computer == 'Rock':
        print('You lose.')
        computer_score = computer_score + 1
    elif player == 'scissors' and computer == 'Paper':
        print('You win!')
        player_score = player_score + 1

#player paper outcomes
    elif player == 'paper' and computer == 'Paper':
        print("It's a draw.")
    elif player == 'paper' and computer == 'Scissors':
        print('You lose.')
        computer_score = computer_score + 1
    elif player == 'paper' and computer == 'Rock':
        print('You win!')
        player_score = player_score + 1
    else:
        print('Invalid input. Please select Rock, Paper, or Scissors')
        continue
        
    x = input('Play Again (yes or no): ')

print()
print('Thanks for playing!')
print(f'Final score: \n Player: {player_score} \n Computer: {computer_score}')
if player_score > computer_score :
    print('Congrats! You win!!!')

if player_score == computer_score :
    print("It's a draw.")

if player_score < computer_score :
    print('Boohoo. You lose :(')
    
```
